# Why Risk Management is Needed?

1. Pure Risk
 
 1. Insurance
 
    - ARIA (American Risk and Insurance Association) : "Insurance is the pooling of fortuitous losses by transfer of such risks to insurers, who agree to indemnify insureds for such losses, to provide other pecuniary benefits on their occurence, or to render services connected with the risk.
    
    - An insured is an individual, household or firm seeking protection from pure risk under and insurance contract.
    
    - An insured seeks to use insurance for its ability to transfer pure risk to the insurer as per the insurance contract.
    
    - The insured does this transfer of pure risk with the hope that the insurer is financially stronger than the insured to bear the loss, and will pay for the loss.
    
    - An insurer pays indemnification to help an insured to restore his/her approximate financial position prior to the occurence of the loss.
    
    - The insured pays a peridic, or in some cases a lump-sum, premium for the purchase of this protection from pure risk.
    
 2. Pooling of Risk Events
    
    - Pooling of losses is the process of spreading of losses incurred due to pure risk events experienced by a few over the entire group.
    
    - The insurer makes a business case by having a large set of customers transfer their pure risk over the insurer, where at a time only a small fraction of the insured will experience a loss event.
    
    - The accumulated premium collected from the large group of customers can be used to indemnify the few who suffer the loss.
    
 3. Insurabile Risk
    
    - This spreading of loss can be done viably by the insurer only if the payment towards indemnification is for fortuitous losses.
    
    - Insurance is meant to pay for fortuitous loss, ones that are unforeseen and unexpected, and occur as a result of chance.
    
    - The Low of Large Numbers is used to model the fair amount of indemnification.
 
2. Speculative Risk

 1. Any investment incurs risk lest it permits arbitrage opportunity, which is the pre-requisite of equilibrium analysis.
 
 2. Making risk management an integral part of the management and control process for running a firm or a household allows for sounder economic management of the enterprise (in the long run).

# Risk Preference vs Risk Aversion

1. Risk Preference 
 1. Preference relation is an ordering for economic outcome, for both non-risky and risky alternatives.

 2. Upon some axioms, preference relation on risk could be represented by a utility function -- cardinal preference ordering.

2. Risk Aversion
  
 1. With cardinal preference ordering, risk aversion is modeled by the concavity of a utility function.

 2. [Jenson's Inequality](http://en.wikipedia.org/wiki/Jensen's_inequality)
   
    - If $\phi(\cdot)$ is a convex function and $X$ is a random variable, then
    
    $$\phi(E[X]) \leq E[\phi(X)]$$
 
 3. Insurance Premium (or Risk Premium)
    - Assume that $U(\cdot)$ is a strictly concave utility function, then the insurance premium for the risk represented by a randome variable $X$, is defined to be $W$ satisfying
    
    $$U(E[X-W]) = E[U(X)]$$
 

# Risk Measures

1. A risk measure is a real-valued mapping on a set of risk factors ($\rho : \tilde{X} \rightarrow R$).

2. Examples of risk measures

 1. Variance or standard deviation of portfolio returns
 
 2. Beta is a risk measure used as an indicator of the impact of systematic risk due to market or a benchmark index portfolio.
 
3. Coherent Risk Measures (See Artzner et al. (1998))

 1. Translation Invariance
 
    $$\rho(\tilde{X} + c) = \rho(\tilde{X}-c) ,$$
    where $c$ is any cashflow.
   
 2. Monotonicity
 
    $$\rho(\tilde{X}_1) \leq \rho(\tilde{X}_2),\quad \mbox{whenever}\; \tilde{X}_1 \geq \tilde{X}_2$$ 
    
 3. Convexity

    For any $\alpha \in [0,1]$,
    $$\rho(\alpha\tilde{X}_1 + (1-\alpha)\tilde{X}_2) \leq \alpha\rho(\tilde{X}_1) + (1-\alpha)\rho(\tilde{X}_2)$$
    
 4. Positive Homogeneity
 
    For any $\alpha > 0$,
    $$\rho(\alpha\tilde{X} = \alpha\rho(\tilde{X})$$
 
4. Value-at-Risk (VaR)

 1. A quantile of risk factor outcome $X$ at a confidence level ($\alpha$)
 
    $$\mbox{VaR}_\alpha = -\inf\{q\in R| P(X\leq q)\geq 1-\alpha) \}$$
 
 2. VaR is not convex
 
    Note that for independent $X, X_1, X_2 \sim N(0,1)$, $X_1 + X_2 \sim N(0,2)$.
    Then we have    
    $$
      P\Big(\frac{X_1+X_2}{\sqrt{2}} = VaR_\alpha(X)\Big) = 1-\alpha ,
    $$
    and 
    $$
      VaR_\alpha(\frac{1}{2}X_1+\frac{1}{2}X_2) = \frac{\sqrt{2}}{2} VaR_\alpha(X) <  \frac{1}{2}VaR_\alpha(X_1)
    $$



5. Conditional VaR

 1. Conditional VaR at a confidence level ($\alpha$)
 
    $$\mbox{CVaR}_\alpha = E\big[X| X<\mbox{VaR}_\alpha\big] $$
 
 2. VaR is not convex
 
    Note that for independent $X, X_1, X_2 \sim N(0,1)$, $X_1 + X_2 \sim N(0,2)$.
    Then we have    
    $$
      P\Big(\frac{X_1+X_2}{\sqrt{2}} = VaR_\alpha(X)\Big) = 1-\alpha ,
    $$
    and 
    $$
      VaR_\alpha(\frac{1}{2}X_1+\frac{1}{2}X_2) = \frac{\sqrt{2}}{2} VaR_\alpha(X) <  \frac{1}{2}VaR_\alpha(X_1)
    $$



# Simulating Measuring Risk

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
n = 1000
mu = 0
sig = 1
dist = st.norm(mu, sig)
data = dist.rvs(n)

fig, ax = plt.subplots()
xmin, xmax = mu - 6 * sig, mu + 6 * sig
ax.set_xlim(xmin, xmax)
ax.hist(data, bins=60, alpha=0.5, normed=True)
xgrid = np.linspace(xmin, xmax, 200)
labelval = r'$N(' + str(mu) + r', ' + `sig*sig` + r')$'
ax.plot(xgrid, dist.pdf(xgrid), 'k-', lw=2, label=labelval)
ax.legend()

plt.show()
```

```python
data.sort()

alpha = 0.95
varind = int(n*(1-alpha))
print '%3.1f%% VaR = %7.4f' % (100*alpha, data[varind-1])
```

```python
alpha = 0.95
varind = int(n*(1-alpha))

print '%3.1f%% CVaR = %7.4f' % (100*alpha, data[:varind].mean())
```

```python
mu = 5
sig = 5
dist = st.norm(mu, sig)
data = dist.rvs(n)

fig, ax = plt.subplots()
xmin, xmax = mu - 6 * sig, mu + 6 * sig
ax.set_xlim(xmin, xmax)
ax.hist(data, bins=60, alpha=0.5, normed=True)
xgrid = np.linspace(xmin, xmax, 200)
labelval = r'$N(' + str(mu) + r', ' + `sig*sig` + r')$'
ax.plot(xgrid, dist.pdf(xgrid), 'k-', lw=2, label=labelval)
ax.legend()

plt.show()
```

```python
data.sort()

alpha = 0.95
varind = int(n*(1-alpha))
print '%3.1f%% VaR = %7.4f' % (100*alpha, data[varind-1])
print '%3.1f%% CVaR = %7.4f' % (100*alpha, data[:varind].mean())

alpha = 0.99
varind = int(n*(1-alpha))
print '%3.1f%% VaR = %7.4f' % (100*alpha, data[varind-1])
print '%3.1f%% CVaR = %7.4f' % (100*alpha, data[:varind].mean())
```

# Get Stock Historical Value from Yahoo Finance (+ Python URLLIB)

- [Python URLLIB Example](http://code.activestate.com/recipes/576495-get-a-stock-historical-value-from-google-finance/)
- [Yahoo Finance Data Sample](http://finance.yahoo.com/q/hp?s=FMI)

```python
import urllib
from datetime import datetime
from threading import Thread
from Queue import Queue

base_url="http://ichart.finance.yahoo.com/table.csv?"

def get_historical(symbols,start=None,end=None,threads=0):
    if isinstance(symbols,str):
        return get_historical_single(symbols,start,end)
    quotes={}
    if threads:
        def quoter(): 
            while True: 
                data = q.get()
                quotes[data[0]]=get_historical_single(data[0],data[1],data[2])
                q.task_done()
        q = Queue() 
        for i in range(threads): 
             t = Thread(target=quoter)
             t.setDaemon(True)
             t.start() 
        for sym in symbols: q.put((sym,start,end))
        q.join()
    else:
        for sym in symbols:
            quotes[sym]=get_historical_single(sym,start,end)
    return quotes

def get_historical_single(symbol,start=None,end=None):
    full_url=base_url+"&s="+symbol
    if start:
        full_url+="&a=%i&b=%i&c=%i"%(start.month-1,start.day,start.year)
    if end:
        full_url+="&d=%i&e=%i&f=%i"%(end.month-1,end.day,end.year)
    full_url+="&g=d"
    quotes={}
    quotes['raw']=[]
    quotes['by_date']={}
    quotes['dates']=[]
    quotes['opens']=[]
    quotes['highs']=[]
    quotes['lows']=[]
    quotes['closes']=[]
    quotes['volumes']=[]
    quotes['adjusted_closes']=[]
    quotes_lines=urllib.urlopen(full_url).read().split('\n')[1:-1]
    for quote_line in quotes_lines:
        #quote_line structure: Date,Open,High,Low,Close,Volume,Adj Close
        splt_q=quote_line.split(',')
        date=datetime(*(map(int,splt_q[0].split('-'))))
        op=float(splt_q[1])
        hi=float(splt_q[2])
        lo=float(splt_q[3])
        close=float(splt_q[4])
        volume=int(splt_q[5])
        adj_close=float(splt_q[6])
        quote=dict(date=date,open=op,high=hi,low=lo,close=close,volume=volume,adj_close=adj_close)
        quotes['raw'].append(quote)
        quotes['by_date'][date]=quote
        quotes['dates'].append(date)
        quotes['opens'].append(op)
        quotes['highs'].append(hi)
        quotes['lows'].append(lo)
        quotes['closes'].append(close)
        quotes['volumes'].append(volume)
        quotes['adjusted_closes'].append(adj_close)
    return quotes

#if __name__ == '__main__':
#    start_date=datetime(2005,1,1)
#    symbols=['F.MI','AAPL','IBM','GOOG']
#    quotes=get_historical(symbols,start=start_date,threads=4)
#    for k in symbols:
#        print '%s: %i quotes'%(k,len(quotes[k]['closes']))
```

```python
start_date=datetime(2015,1,1)
#symbols = ['F.MI','AAPL','IBM','GOOG']
symbols = ['FMI','AAPL','IBM','GOOG'] # FMI : Foundation Medicine ??
quotes=get_historical(symbols,start=start_date,threads=4)
for k in symbols:
    print '%s: %i quotes'%(k,len(quotes[k]['closes']))
```
