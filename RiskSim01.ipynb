## Types of Risk

1. Pure Risk

 Variability quantified in terms of probabilities that can either result in realization of losses or no losses.
 
2. Speculative Risk

 Variability quantified in terms of probabilities that can result in realization of both gains and losses

## Getting Started with Modeling Risk

### Random Variable and Probability
  - Probability space : $(\Omega, \mathcal{F}, P)$
  - Random variable : $X: \Omega \rightarrow R$
  - Cummulative distribution function : $F(x) := P(X(\omega))\leq x)$
  - Probability density function : $f(x)$ given with 
  $$F(x) = \int_{-\infty}^x f(y)dy$$
  or 
  $$F(x) = \sum_{x_i< x} f(x_i)$$
  - Expectation : 
  $$E[g(X)] = \int_{-\infty}^\infty g(y) f(y)dy$$ 
  or
  $$E[g(X)] = \sum_{x_i} g(x_i) f(x_i)$$
  - Joint distribution : 
  $$F(x_1, x_2) = \int_{-\infty}^{x_1}\int_{-\infty}^{x_2}f(y_1,y_2)dy_1dy_2$$
  or
  $$F(x_1, x_2) = \sum_{x_1(i)< x_1}\sum_{x_2(j)< x_2}f(x_1(i),x_2(j))$$
  - Marginal distribution :
  $$F_{X_1}(x_1) = \int_{-\infty}^{x_1}\int_{-\infty}^{\infty}f(y_1,y_2)dy_1dy_2$$
  or
  $$F_{X_1}(x_1) = \sum_{x_1(i)< x_1}\sum_{x_2(j)}f(x_1(i),x_2(j))$$
  - Conditional distribution :
  $$F_{X_2}(x_2 | X_1 = x_1) = \int_{-\infty}^{x_2} f_{X_2}(x_2 | X_1 = x_1) dx_2 ,$$
  where
  $$f_{X_2}(x_2 | X_1 = x_1) = \frac{f(x,y)}{f_{X_1}(x_1)} $$

Note that scipy.stats.norm() generates an object abstacting normal distribution. It is a member, pdf() to calculate the probability density function of that distribution.

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

x = np.arange(-5,5,0.1)
dist1 = st.norm(0, 1)
y = dist1.pdf(x)
plt.plot(x,y,'r--')
x = 2*x
dist2 = st.norm(0, 2**2)
y = dist2.pdf(x)
plt.plot(x,y,'b+')
````

### Limit Theorems

A nicely summarized introduction to Law of Large Numbers (LLT) and Central Limit Theorem (CLT) could be found on ['LLT and CLT' of Quant-Econ.net](http://quant-econ.net/py/lln_clt.html)

Let us define sample mean $\bar{X}_n :=\frac{1}{n}\sum_{i=1}^n X_i$. Then,
1. Law of Large Numbers
The Classical Version
$$
\bar{X}_n \longrightarrow \mu\quad \mbox{a.s.},
$$
or
$$
P\big(\bar{X}_n \rightarrow \mu\quad \mbox{as}\quad n\rightarrow \infty\big) = 1 
$$

A Weaker Version
$$
P\big(|\bar{X}_n-\mu|\geq \epsilon\big) \rightarrow 0\quad
\mbox{as}\quad n\rightarrow\infty .
$$

2. Central Limit Theorem
$$
\sqrt{n} (\bar{X}_n - \mu) \xrightarrow{d} N(0,\sigma^2)\quad \mbox{as}\quad n\rightarrow \infty ,
$$
$$
\bar{X}_n \xrightarrow{d} N(\mu,\frac{\sigma^2}{n})\quad \mbox{as}\quad n\rightarrow \infty .
$$

Exercise : LLT with Normal Distributions

````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

n = 100

# == A collection of distributions == #
distributions = {"N(-10, 1)": norm(-10, 1),
                 "N(0, 1)": norm(0, 1),
                 "N(10, 1)": norm(10, 1)}

# == Create a figure and some axes == #
num_plots = 3
fig, axes = plt.subplots(num_plots, 1, figsize=(10, 10))

# == Set some plotting parameters to improve layout == #
bbox = (0., 1.02, 1., .102)
legend_args = {'ncol': 2,
               'bbox_to_anchor': bbox,
               'loc': 3,
               'mode': 'expand'}
plt.subplots_adjust(hspace=0.5)

keys = list(distributions.keys())
for ax, key in zip(axes, keys):
    # == Generate n draws from the distribution == #
    distribution = distributions[key]
    data = distribution.rvs(n)

    # == Compute sample mean at each n == #
    sample_mean = np.empty(n)
    for i in range(n):
        sample_mean[i] = np.mean(data[:i+1])

    # == Plot == #
    ax.plot(list(range(n)), data, 'o', color='grey', alpha=0.5)
    axlabel = r'$\bar X_n$' + ' for ' + r'$X_i \sim$' + ' ' + key    
    ax.plot(list(range(n)), sample_mean, 'g-', lw=3, alpha=0.6, label=axlabel)
    m = distribution.mean()
    ax.plot(list(range(n)), [m] * n, 'k--', lw=1.5, label=r'$\mu$')
    ax.vlines(list(range(n)), m, data, lw=0.2)
    ax.legend(**legend_args)
````

Exercise : CLT with the standart normal distribution

````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
from matplotlib import rc

# == Specifying font, needs LaTeX integration == #
rc('font', **{'family': 'serif', 'serif': ['Palatino']})
rc('text', usetex=True)

# == Set parameters == #
n = 250     # Choice of n
k = 1000  # Number of draws of Y_n
mu = 0
s = 1
distribution = norm(mu, s)  # Standard normal distribution

# == Draw underlying RVs. Each row contains a draw of X_1,..,X_n == #
data = distribution.rvs((k, n))
# == Compute mean of each row, producing k draws of \bar X_n == #
sample_means = data.mean(axis=1)
# == Generate observations of Y_n == #
Y = np.sqrt(n) * (sample_means - mu)

# == Plot == #
fig, ax = plt.subplots()
xmin, xmax = -3 * s, 3 * s
ax.set_xlim(xmin, xmax)
ax.hist(Y, bins=60, alpha=0.5, normed=True)
xgrid = np.linspace(xmin, xmax, 200)
ax.plot(xgrid, norm.pdf(xgrid, scale=s), 'k-', lw=2, label=r'$N(0, \sigma^2)$')
ax.legend()

plt.show()
````
